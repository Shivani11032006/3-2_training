```python
#Class with Constructor and Methods
class Register:
    # Constructor
    def __init__(self, reg_no, name, course):
        self.reg_no = reg_no
        self.name = name
        self.course = course

    # Method to display details
    def display(self):
        print("Register Number :", self.reg_no)
        print("Student Name    :", self.name)
        print("Course          :", self.course)
        print("----------------------------")

    # Method to update course
    def update_course(self, new_course):
        self.course = new_course
```

```python
#Function Outside the Class
def create_register():
    reg_no = int(input("Enter Register Number: "))
    name = input("Enter Student Name: ")
    course = input("Enter Course: ")

    student = Register(reg_no, name, course)
    return student
```

```python
# Creating a register entry
student1 = create_register()

# Display register details
student1.display()

# Update course
new_course = input("Enter new course to update: ")
student1.update_course(new_course)

print("\nUpdated Register Details:")
student1.display()
```
